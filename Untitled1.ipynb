```python
from numpy import pi, sin, cos, mgrid
from scipy.special import jn, jn_zeros
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as pl
import numpy as np
from scipy.special import jv, jn_zeros
```

```python
RADIUS = 1
SPEED_OF_SOUND = 0.75
BESSEL_ROOTS = [jn_zeros(m, 10) for m in range(10)]

def lambda_mn(m, n, radius):
    return BESSEL_ROOTS[m][n - 1] / radius


def circular_membrane(r, theta, m, n, radius, speed_of_sound):
    l = lambda_mn(m, n, radius)
    R = jv(m, l * r)
    Theta = np.cos(m * theta)
    return R * Theta

r = np.linspace(0, RADIUS, 100)
theta = np.linspace(0, 2 * np.pi, 100)

r, theta = np.meshgrid(r, theta)
x = np.cos(theta) * r
y = np.sin(theta) * r
```

```python
m, n = 1,2
fig = pl.figure(figsize=(10, 5))
ax = fig.add_axes([0, 0, 1, 1], projection='3d')
ax.set_axis_off()
z = circular_membrane(r, theta, m, n, RADIUS, SPEED_OF_SOUND)
vmax = np.max(jv(m, np.linspace(0, BESSEL_ROOTS[m][n], 100)))
#ax.plot_surface(x, y, z, linewidth=0, cmap='Spectral', vmin=-vmax, vmax=vmax, rcount=100, ccount=100)
#ax.plot_wireframe(x, y, z)
ax.plot_surface(x, y, z, cmap=pl.cm.YlGnBu_r)
ax.set_zlim(-1.1, 1.1)
ax.set_xlim(-0.75, 0.75)
ax.set_ylim(-0.75, 0.75)
omega = SPEED_OF_SOUND * lambda_mn(m, n, RADIUS)
ax.set_title(
        f'Circular membrane, m = {m}, n = {n}, Ï‰={omega:.2f}',
        size=14
    )
```
