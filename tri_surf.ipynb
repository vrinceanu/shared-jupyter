```python
import numpy as np, matplotlib.pyplot as pl
fig = pl.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1, projection='3d')

A = np.array((2,3,3))
B = np.array((4,3,1))
C = np.array((3,1,3))
D = B + C - A
Quadrangle=np.array((A,B,C,D))
names = ("A","B","C","D")
ax.plot_trisurf(Quadrangle[:,0],Quadrangle[:,1],triangles=((0,1,2),(1,3,2)),Z=Quadrangle[:,2])
ax.set_xlim(0,5)
ax.set_ylim(0,5)
ax.set_zlim(0,5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
for i in range(len(names)):
    ax.text(Quadrangle[i,0],Quadrangle[i,1],Quadrangle[i,2],names[i])
ax.view_init(30,-70)
```
